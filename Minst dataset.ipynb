```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten,BatchNormalization,MaxPooling2D,AveragePooling2D,Dropout
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[10])
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
#create model Where 3 layers
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28),name="input_layer"))
model.add(Dense(700,activation = "relu",name="hidden_layer_1"))
# model.add(MaxPooling2D(pool_size=(2,2),strides=2))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))
model.add(Dense(350,activation = "relu",name="hidden_layer_2"))
# model.add(Dropout(0.5))
# model.add(MaxPooling2D(pool_size=(2,2),strides=2),)
# model.add(Dense(32,activation = "relu",name="hidden_layer_5"))
# model.add(Dropout(0.5))
# model.add(BatchNormalization())
model.add(Dense(10,activation = "softmax",name="hidden_layer_3"))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='Adam',
              metrics=['accuracy'])
 #String (name of optimizer) or optimizer instance. See tf.keras.optimizers.
```

```python
history = model.fit(X_train,y_train,epochs=10,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
```

```python
plt.plot(history.history["accuracy"])
plt.label="train_accuracy"
plt.plot(history.history["val_accuracy"],label="test_accuracy")
```
